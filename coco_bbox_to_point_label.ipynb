```python
import json
import os
```

```python
def convert_bbox_to_point_label(input_dir):
    with open(os.path.join(input_dir, "labels.json")) as f:
        data = json.load(f)

    for ann in data['annotations']:
        x = ann['bbox'][0]
        y = ann['bbox'][1]
        w = ann['bbox'][2]
        h = ann['bbox'][3]

        xc = round(x + (w/2), 3)
        yc = round(y + (h/2), 3)

        # print(x, y, w, h, xc, yc)
        ann['point'] = [xc, yc]
        del ann['bbox']

    with open(os.path.join(input_dir, "point_labels.json"), 'w') as f:
        json.dump(data, f)
```
